<p><font color="blue">Paso por referencia: La variable pasa directo a la función, y el objeto (El contenido de la variable) implicita se encuentra ahí. El argumento (objeto) sólo es un alias para la variable cuando lo llaman. Por lo tanto es el mismo objeto en el mismo lugar de la memoria.
Paso por valor: La función recibe una copia del objeto cuando lo llaman y por lo tanto tiene una nueva locación en la memoria. Los objetos podrán tener el mismo valor, pero estarán en diferentes ubicaciones por lo que lo que le pase a uno, no afectará al otro.
Python maneja la funcionalidad de "Los objetos de referencia son llamados por valor" donde el mismo objeto crea una variable para si misma. El mismo objeto puede ser llamado por diferentes variables https://robertheaton.com/2014/02/09/pythons-pass-by-object-reference-as-explained-by-philip-k-dick/
 </font></p>

<p>La herencia es programación es cuando una clase utiliza código que está dentro de otro código. Las subclases o subclases hijos,heredan variables y métodos de las clases padres. 
Python utiliza dos tipos de herencia:
<ul>
    <li type="circle"><b>Herencia Simple:</b> La herencia simple puede definirse como los métodos padre e hijo donde como ya se ha especificado el hijo mantiene las variables y métodos de las clase padre, en Python se pueden reescribir los clase padre y seguir manteniendo algunas varibles o métodos originales con la función .super() . </li>
    <li type="circle"><b>Herencia Múltiple:</b> La herencia múltiple es cuando las clase hijos reciben variables o métodos de más de una clase padre.</li>
</ul>
