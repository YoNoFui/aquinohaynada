# Heap y Stack en memoria, son para todos los procesos
##  - Stack: lo que se declaró en el programa de tamaño fijo, inclusive un ciclo que tiene un tamaño definido (   for (int i =0;i<10)   )
##  - Heap: Todo lo que se declara que no tiene un tamaño fija: lista, for elemento in lista, malloc, calloc, TODOS los iteradores viven en Heap

## Itertools reserva un espacio en el Heap solo para Python (lo cual no se permite normalmente), por eso ya no necesita control de memoria y usa la evaluación perezosa (lazzy).
## Los beneficios son en tiempo de ejecución, pero tiene un impacto en recursos.

### Iteradores infinitos.
#### Un iterador es un conjunto de elementos que tinen un indicador(índice)
#### Los iteradores indefinidos, infinitos: count(), cycle(), repeat(), no deben ir solos

# Cycle(): Puede iterar sobre una lista muy larga y está optimizado para esas aplicaciones.

```python
contador = 0

for item in cycle([1,2,3]):
    if contador ==5:
        break
        
    contador +=1
    
    print(item)
```

# Pickle
### Archivo de tipo binario serializable 
### Se puede guardar cualqueir contenido de una variable sin pérdida de datos ya que es al mas bajo nivel.

```python
import pickle
```

```python
favorite_color = {"lion":"yellow","kitty":"fred"}
```

### Se guarda la variable en el archivo "Save.p" "wb"-> write binary

```python
pickle.dump(favorite_color, open("Save.p","wb"))
```

### Se recupera la variable del archivo "Save.p" "rb"-> read binary

```python
favorite_color = pickle.load(open("Save.p","rb"))
```

```python
favorite_color
```
